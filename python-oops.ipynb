```python
class Student:
  # class variable, accessible and shared by all obj instances
  total_students = 0
  mean_average = 0
  total_marks = 0

  def __init__(self, name, email, marks):
    # self means the object instance
    self.name=name
    self.email=email
    self.marks=marks
    Student.total_students+=1
    Student.update_mean_average(self.marks)

  def getID(self):
    return self.name+'_'+self.email

  @classmethod
  def get_total_students(cls):
    return cls.total_students

  @classmethod
  def see(cls):
    return cls

  @staticmethod
  def update_mean_average(marks):
    # these are helper methods which can be shared by all member methods
    Student.total_marks+=marks;
    Student.mean_average=Student.total_marks/Student.total_students

  @classmethod
  def get_mean_average(cls):
    return cls.mean_average

  def isStudentAboveAverage(self):
    return self.marks>Student.mean_average

s1 = Student('Tunir1','sahatunir1@gmail.com', 50)
print(s1.getID(), s1.get_total_students())

s2 = Student('Tunir2','sahatunir2@gmail.com', 90)
print(s2.getID(), s2.get_total_students())

s3 = Student('Tunir3','sahatunir3@gmail.com', 10)
print(s3.getID(), s3.get_total_students())

print(s1.get_total_students(),s2.get_total_students(),s3.get_total_students())
print(s1.see())
print(s1.isStudentAboveAverage(), s2.isStudentAboveAverage(), s3.isStudentAboveAverage())
```

```python
#Inheritance

class Teacher(Student):

  def __init__(self, name, email, marks, subject):
    super().__init__(name, email, marks)
    self.subject=subject


t1 = Teacher('Tunir4', 'sahatunir4@gmail.com', 100, 'Math')
print(t1.getID(), t1.subject)
```
